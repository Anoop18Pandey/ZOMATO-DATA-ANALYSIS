ZOMZTO DATA ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv('Zomato data .csv')
dataset.info()
```

```python
def handlerate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
dataset['rate']  = dataset['rate'].apply(handlerate)
dataset.head()
```

Q1. TYPE OF RESTURANT

```python
sns.countplot(x=dataset['listed_in(type)'])
plt.xlabel('type of Resturant')
plt.savefig('type_of_Resturant.png')
```

conclusion: mejority of the resturant falls in dining category

Q2. TOTAL VOTES OF EACH TYPE OF RESTURANT HAVE

```python
grouped_data = dataset.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, color='green', marker='o')
plt.xlabel('type of resturant', color='red',size=20)
plt.ylabel('votes', color='red', size=20)
plt.savefig('votes_each_resturant.png')
```

conclusion: dining resturant has receved the most votes

Q3. RETTING OF THE RESTURANT

```python
plt.hist(dataset['rate'],bins=5)
plt.title('rating distribution')
plt.show()
plt.savefig('rating_distribution.png')
```

conclusion: the mejority resturant receved rating 3.5 to 4.5

Q4. AVERAGE ONLINE SPENDING BY THE COUPLE 

```python
couple_data = dataset['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.savefig('average_spend.png')
```

conclusion: 300 is the most time spent by the couple in onle order 

Q5. WHICH MODE HAS THE MOST RATING

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
dataset = pd.read_csv('Zomato data .csv')
plt.Figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate', data=dataset)
plt.savefig('mode_most_rating.png')
```

conclusion: online order receved the more rating

Q6. WHICH TYPE OF THE RESTURANT RECEVED THE OFFLINE ORDER MOST

```python
pivot_table = dataset.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('online_order')
plt.ylabel('listed_in(type)')
plt.show()
plt.savefig('resturant receved online orde.png')
```

conclusion: dining resturant has most online and offline order
